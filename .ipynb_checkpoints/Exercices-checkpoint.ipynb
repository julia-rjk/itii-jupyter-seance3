```python
import numpy as np

# Rappel Soit f une fonction continue d√©finie sur un intervalle I et, a et b deux r√©els de I.
# Pour tout r√©el k compris entre f(a) et f(b), il existe au moins un r√©el c compris entre a et b tel que f(c)=k.
# Soit f une fonction continue et strictement monotone sur un intervalle [a;b]
# Pour tout nombre k compris entre f(a) et f(b), l'√©quation f(x)=kadmet une unique solution dans l'intervalle [a;b].
# f(x)=x^3+2x‚àí2

# v√©rifier continuit√© et stricte monotonie de f sur [0;1]
def verifier(a, b):
    print(" f(a) : " , f(a)) # -2
    print(" f(b) : " ,f(b)) # 1
    c = 0.1
    while(c < 1):
        k = f(c)
        if(k <= f(b) and k >= f(a)):
            print("Valeur de k :", k)
            print("Valeur de c :" , c)
            break;
        else:
            c = c + 0.1
    
def f(x):
    return pow(x, 3) + 2*x - 2


def verifierMonotonie():
    c = 0
    last = 0 
    croissante = 'true'
    while(c < 1):
        k = f(c)
        if(k < f(last)):
            croissante = 'false'
            break
        else:
            c = c + 0.01
            last = k
    if(croissante):
        print("La fonction est strictement croissante")
    else: 
        print("La fonction est strictement d√©croissante")

def main():
    # v√©rifier continuit√©
    verifier(0,1)
    # v√©rifier monotonie
    verifierMonotonie()
    
    # Pour savoir si x0 se trouve dans [0;0,5]ou dans [0,5;1], il faut v√©rifier si f(0) et f(0,5) sont de m√™mes signes

    if(f(0) > 0 and f(0.5) > 0):
        print("f(0) et f(0,5) : m√™me signe positif")
    elif (f(0) < 0 and f(0.5) < 0):
        print("f(0) et f(0,5) : m√™me signe n√©gatif")
    else:
        print("f(0) et f(0,5) : pas le m√™me signe")

# proposer un algorithme de recherche pour √©valuer x_0 en fonction de ùëÅ , donner par l'utilisateur

main()
```
